```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Weather Data
weather_data = pd.read_csv('/content/GlobalWeatherRepository.csv')

# Step 2: Preprocess the Data
# Drop any rows with missing values
weather_data.dropna(inplace=True)

# Exclude non-numeric columns from the scaling process
numeric_columns = weather_data.select_dtypes(include=np.number).columns
scaler = StandardScaler()
scaled_features = scaler.fit_transform(weather_data[numeric_columns])

# Step 3: Choose the Number of Clusters (K)
# Use the elbow method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), wcss)
plt.title('Global Weather')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Based on the elbow curve, choose the optimal number of clusters
k = 3

# Step 4: Apply K-means Clustering
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(scaled_features)

# Step 5: Visualize the Clusters
# For example, plot the clusters in a scatter plot
plt.scatter(weather_data['latitude'],weather_data['longitude'], c=clusters, cmap='viridis')
plt.title('Weather Data Clusters')
plt.xlabel('41.33')
plt.ylabel('19.82')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
dates = ['2023-08-29', '2024-02-23']
temperatures = [28.8, 14.0]  # Temperatures in Celsius
precipitations = [0, 5,]  # Precipitation in mm

# Create figure and axis objects
fig, ax1 = plt.subplots()

# Plot temperature data
ax1.plot(dates, temperatures, color='tab:red')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.bar(dates, precipitations, color='tab:blue', alpha=0.5)
ax2.set_ylabel('Precipitation (mm)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Add title and show plot
plt.title('Weather Forecast')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
countries = ['Albania', 'Algeria', 'Andorra',]
temperatures = [27, 28, 10.2,]  # Temperatures in Celsius
humidity = [60, 65, 70, 75, ]  # Humidity in percentage

print(f"Length of countries: {len(countries)}")
print(f"Length of temperatures: {len(temperatures)}")
print(f"Length of humidity: {len(humidity)}")

# Adjust the length of the shortest list to match the others
if len(countries) == min(len(countries), len(temperatures), len(humidity)):
    temperatures = temperatures[:len(countries)]
    humidity = humidity[:len(countries)]
elif len(temperatures) == min(len(countries), len(temperatures), len(humidity)):
    countries = countries[:len(temperatures)]
    humidity = humidity[:len(temperatures)]
else:
    countries = countries[:len(humidity)]
    temperatures = temperatures[:len(humidity)]

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(temperatures, humidity, marker='o', color='blue')

# Add labels and title
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.title('Temperature vs Humidity')

# Add country labels
for i, country in enumerate(countries):
    plt.text(temperatures[i], humidity[i], country)

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
countries = ['Albania', 'Algeria', 'Andorra',]
temperatures = [27, 28, 10.2,]  # Temperatures in Celsius
wind_pressure = [6.4, 0.5, 139.6,]

# Plotting
plt.figure(figsize=(10, 6))

# Plot temperature
plt.plot(countries, temperatures, marker='o', label='Temperature (°C)', color='blue')

# Plot wind pressure
plt.plot(countries, wind_pressure, marker='s', label='Wind Pressure (mb)', color='red')

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Value')
plt.title('Temperature and Wind Pressure by Country')
plt.xticks(rotation=45)  # Rotate country names for better readability
plt.grid(True)

# Adding legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
