```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the economics data
data = pd.read_csv('/content/diffusion_index_in_japan.csv')

# Select features for clustering
X = data[['HouseholdTrend', 'HouseholdTrend_Retail']]
Y = data[['month', 'day']]

# Standardize the features
X = (X - X.mean()) / X.std()

# Perform K-means clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Add cluster labels to the DataFrame
data['Cluster'] = kmeans.labels_

# Plot the clusters
plt.figure(figsize=(8, 6))

# Perform K-means clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Add cluster labels to the DataFrame
data['Cluster'] = kmeans.labels_

# Plot the clusters
plt.figure(figsize=(8, 6))

# Plot each cluster
for cluster in data['Cluster'].unique():
    plt.scatter(data[data['Cluster'] == cluster]['HouseholdTrend'],
                data[data['Cluster'] == cluster]['month'],
                label=f'Cluster {cluster}')

plt.xlabel('HouseholdTrend')
plt.ylabel('month')
plt.title('Clusters of Economics Data')
plt.legend()
plt.show()
```
